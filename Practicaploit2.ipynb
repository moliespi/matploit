Ejercicio 1: Agrupa los siguientes datos en intervalos de amplitud 2:
2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6
Construye una tabla de frecuencias completa y dibuja el histograma
correspondiente.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = [2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6]

bins = np.arange(min(data) -1, max(data) + 3, 2)

#Agruparlo en intervalos
binned_data = pd.cut(data, bins=bins, right=False, labels=bins[:-1])

# Create frequency table
tabla_frecuencia = pd.DataFrame({'Intervalos': bins[:-1], 'Frecuencia': binned_data.value_counts().sort_index()})
display(tabla_frecuencia)

# Draw histogram
plt.hist(data, bins=bins, rwidth=1, color='red')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histograma con intervalos de amplitud 2')
plt.xticks(bins[:-1] + 1, [f'{bins[i]}-{bins[i+1]-1}' for i in range(len(bins)-1)]) # Label with intervals
plt.show()
```

5. El número de hermanos de los alumnos de una clase está reflejado en los
siguientes datos:
1, 2, 2, 3, 0, 1, 2, 1, 3, 0, 0, 2, 1, 1, 4, 3, 4, 2, 1, 4, 3, 2, 1, 0, 1, 0, 1, 1, 2, 1, 0, 3
Construye una tabla con frecuencias absolutas y porcentajes. Crea un diagrama
de sectores basándote en los porcentajes.

```python
data = [1, 2, 2, 3, 0, 1, 2, 1, 3, 0, 0, 2, 1, 1, 4, 3, 4, 2, 1, 4, 3, 2, 1, 0, 1, 0, 1, 1, 2, 1, 0, 3 ]

bins = np.arange(min(data), max(data) + 2, 1)

binned_data = pd.cut(data, bins=bins, right=False, labels=bins[:-1])

# Create frequency table
tabla_frecuencia = pd.DataFrame({'Cant_bro': bins[:-1], 'Alumnos': binned_data.value_counts().sort_index()})
tabla_frecuencia['Porcentaje'] = (tabla_frecuencia['Alumnos'] / tabla_frecuencia['Alumnos'].sum()) * 100

print("Tabla de frecuencias")
display(tabla_frecuencia)

# Draw histogram
fig, ax = plt.subplots()
plt.figure(figsize=(5, 5))
ax.bar(tabla_frecuencia['Cant_bro'], tabla_frecuencia['Alumnos'], color='purple')
ax.set_xlabel('Cantidad de hermanos')
ax.set_ylabel('Alumnos')
ax.set_title('Número de hermanos')
plt.show()
```

7. Se ha realizado una encuesta a 600 chicos y chicas sobre su deporte preferido,

o Baloncesto: 12%

8. La tabla recoge la edad de un grupo de jóvenes encuestados:

9. En el estudio estadístico realizado en un instituto se han obtenido los siguientes
datos:
o Peso (kg): [50, 55), [55, 60), [60, 65), [65, 70), [70, 75]
o Número de alumnos: 10, 40, 25, 20, 5
a) Organiza una tabla estadística. b) Construye el histograma y el
polígono de frecuencias.

```python
peso = [[50, 55], [55, 60], [60, 65], [65, 70], [70, 75]]
num_alum = [10, 40, 25, 20, 5]
data = pd.DataFrame({'Peso (Kg)': peso, 'Numero de alumnos': num_alum})
display(data)

midpoints = [(interval[0] + interval[1]) / 2 for interval in peso]

#histograma
plt.figure(figsize=(8, 6))
bins = [50, 55, 60, 65, 70, 75]
plt.hist(midpoints, bins=bins, weights=num_alum, color='yellow')
plt.xlabel('Peso (Kg)')
plt.ylabel('Numero de alumnos')
plt.title('Histograma de pesos en el instituto')
plt.xticks(bins)
plt.show()

#Poligono de frecuencias
plt.figure(figsize=(8, 6))
plt.plot(midpoints, num_alum, marker='o', linestyle='-', color='green')
plt.xlabel('Peso (Kg)')
plt.ylabel('Numero de alumnos')
plt.title('Polígono de frecuencias de pesos en el instituto')
plt.xticks(midpoints, [f'{i[0]}-{i[1]}' for i in peso])
plt.grid(True)
plt.show()
```

fueron:

15. La distribución del tiempo libre semanal de un grupo de personas es la
siguiente:
o Deporte: 30%
o Lectura: 20%
o Cine: 15%
o Viajes: 10%
o Otras actividades: 25%
a) Representa estos datos mediante un diagrama de sectores.
b) Calcula el ángulo correspondiente a cada sector.

```python
activities = ['Deporte', 'Lectura', 'Cine', 'Viajes', 'Otras actividades']
percentages = [30, 20, 15, 10, 25]

# a) Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=activities, autopct='%1.1f%%', startangle=140)
plt.title('Distribución del tiempo libre semanal')
plt.axis('equal')
plt.show()

# b) Calculate the angle for each sector
total_angle = 360
angles = [(p / 100) * total_angle for p in percentages]

# Display the angles
angles_df = pd.DataFrame({'Actividad': activities, 'Porcentaje': percentages, 'Ángulo (grados)': angles})
display(angles_df)
```
